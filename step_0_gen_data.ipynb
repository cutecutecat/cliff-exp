```python
# generate error test suite
random.seed(0)
np.random.seed(0)

import shutil
from itertools import product

import pandas as pd
from editdistance import eval

if not os.path.exists("Error"):
    os.mkdir("Error")
shutil.copyfile("Correct/R0L7E5V0.csv", "Error/R0D0I0.csv")
shutil.copyfile("Correct/R3L7E5V0.csv", "Error/R3D0I0.csv")
shutil.copyfile("Correct/R6L7E5V0.csv", "Error/R6D0I0.csv")

def select_mutation_seq(src: str, tgt: str) -> bool:
    """Select sequence by edit distance from wild type
    
    Distance 0   | 1   | 2   | 3   | >=4
    
    Drop     0.0 | 0.0 | 0.5 | 0.9 | 0.99
    """
    drop_table_distance = {0: 0.0, 1: 0.0, 2: 0.5, 3: 0.9}
    dis = eval(src, tgt)
    if dis in drop_table_distance:
        return random.random() > drop_table_distance[dis]
    else:
        return random.random() > 0.99

drop_precent = [0.1, 0.3, 0.5, 0.7, 0.9]

files = ["Error/R0D0I0.csv", "Error/R3D0I0.csv", "Error/R6D0I0.csv"]
for f in files:
    df = pd.read_csv(f)
    wild: str = df.loc[0, 'Sequence']
    for drop, index in product(drop_precent, range(INSTANCE)):
        remove_n = int(len(df.index) * drop)
        drop_indices = np.random.choice(df.index, remove_n, replace=False)
        df_subset: pd.DataFrame = df.drop(drop_indices)
        df_subset.to_csv(f"Error/R{f[-9]}D{int(10 * drop)}I{index}.csv")
    
    for index in range(INSTANCE):
        df_subset = df.loc[df['Sequence'].apply(select_mutation_seq, args=(wild, )) == True] 
        df_subset.to_csv(f"Error/R{f[-9]}DSI{index}.csv")
```
